# <span style="color:red">1- Data</span>

```python
import pandas as pd

data  = pd.read_csv("D:/data.csv").values

data.shape
```

# <span style="color:red">2- K_number Selection</span>

```python
from sklearn.cluster import KMeans
```

```python
inertia = []

for k in range(1,6):
    km = KMeans(random_state = 4, n_clusters = k)
    km.fit(data)
    inertia.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.style.use('dark_background')
plt.plot   (range(1,6), inertia)
plt.scatter(range(1,6), inertia, marker='o', c='red', s=200)
print('                  Elbow ====>>  \U0001F9BE')


plt.title('Elbow', size=25, c='red')
plt.xlabel('clusters', size=15)
plt.ylabel('inertia', size=15)
plt.show()
```

 
### <span style="color:green">It turns out that 3 clusters is adequate</span>

# <span style="color:red"> 3-             label (clusters) generation</span>

```python
km = KMeans(random_state = 4, n_clusters = 3)

label = km.fit_predict(data)
label.shape
```

```python
colors = ['r', 'g', 'b']

plt.style.use('dark_background')
for i,co in enumerate(colors):
    plt.scatter(data[label==i, 0], data[label==i, 1], c=co, s=7, label=[f"cluster{i+1}"])
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], c='y', s=200)
    
    
plt.title('Clusters of customers', size=20, c='yellow')
plt.xlabel('Annual Income (k$)', size=12)
plt.ylabel('Spending Score (1-100)', size=12)


plt.legend()
plt.show()
```
